A famous Pizza chain is thinking of expanding it's business in Haryana by opening up Pizza serving places in Haryana.

### How the Problem can be approached ?

To solve the business problem for the pizza company, following solution is proposed
1) Data pertaining to pizza places operating in various districts of the state will be analyzed
2) Areas where there are no pizza places already will be found as well as places with low desnity of pizza places
3) Data will be merged with population density and area of the place to find optimal places for opening a pizza cafe

```python
#K-means clustering
cluster = 3 

#Dataframe for clustering
hry_clustering = hry_pizza.drop(['district'], 1)

#run K-means clustering
k_means = KMeans(init="k-means++", n_clusters=cluster, n_init=12).fit(hry_clustering)

#getting the labels for first 10 locality 
print(k_means.labels_[0:10])
```

```python
#Creating a copy of hry_pizza dataframe
hry_labels = hry_pizza.copy()

#adding label to hry_labels
hry_labels["Cluster Label"] = k_means.labels_

hry_labels.head()
```

```python
#Merging the hry_labels and dd dataframes to get the latitude and longitudes for each district
hry_labels = hry_labels.join(dd.set_index('district'), on = 'district')
hry_labels.head()
```

```python
#Grouping the districts according to their Cluster Labels
hry_labels.sort_values(["Cluster Label"], inplace=True)
hry_labels.head()
```

```python
#Cleaning the dataframe for mapping the districts according to their cluster labels
hry_only_labels = hry_labels.drop(columns=['Pizza Place','latitude','longitude'])
hry_only_labels.head()
```
